```python
"""
使用 Hugging Face Transformers 套件載入指定的語言模型與分詞器 (tokenizer)。

1.指定要載入的模型 ID："LiquidAI/LFM2-350M"
2.AutoModelForCausalLM.from_pretrained設定模型運行環境參數
3.使用 AutoTokenizer.from_pretrained() 載入對應的分詞器，以便將文字轉換為模型可讀取的值。
"""

from transformers import AutoModelForCausalLM, AutoTokenizer

model_id = "LiquidAI/LFM2-350M"
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype="bfloat16",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
# 定義使用者輸入的對話
prompt = "請問高血壓的護理指導重點有哪些？請列點說明"

# 將提示詞轉換成模型輸入值
input_ids = tokenizer.apply_chat_template(
    [{"role": "user", "content": prompt}],
    add_generation_prompt=True,
    return_tensors="pt",
    tokenize=True,
).to(model.device)

# 呼叫模型進行文字生成
output = model.generate(
    input_ids,
    do_sample=True,
    temperature=0.3,
    min_p=0.15,
    repetition_penalty=1.05,
    max_new_tokens=512,
)

print(tokenizer.decode(output[0], skip_special_tokens=False))
```
